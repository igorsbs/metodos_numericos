# Introdução



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Equações algébricas 

```python
def eq_1(x):
    # Função avaliada entre [5,10]
    return -0.5*x**2 + 2.5*x + 4.5

def eq_2(x):
    # Função avaliada entre [0,1]
    return 5*x**3 - 5**x**2 + 6*x - 2

def eq_3(x):
    # Função avaliada entre [0.5,1]
    return -25 + 82*x - 90*x**2 + 44*x**3 - 8*x**4 + 0.7*x**5
```

## Equações transcendentais

```python
def eq_4(x):
    # Função avaliada entre [0.5,1]
    return np.sin(x) - x**3

def eq_5(x):
    # Função avaliada entre [0.5,2]
    return np.log(x**4) - 0.7
```

### Plot das funções

```python
list_of_eqs = [eq_1,eq_2,eq_3,eq_4,eq_5]

ranges_dict = {
    'eq_1':{'xi':5,'xs':10},
    'eq_2':{'xi':0,'xs':1},
    'eq_3':{'xi':0.5,'xs':1},
    'eq_4':{'xi':0.5,'xs':1},
    'eq_5':{'xi':0.5,'xs':2}
    }
```

```python



plt.figure(figsize=(12,6))
i = 1
for key, _ in ranges_dict.items():
    x = np.arange(ranges_dict[key]['xi'],ranges_dict[key]['xs']+0.01,0.01)    
    plt.subplot(1,5,i)
    plt.plot(x,[list_of_eqs[i-1](x_i) for x_i in x])
    plt.plot(x,[0 for _ in x])
    plt.title(key)
    plt.grid(True)
    plt.xlabel('x')
    if i == 1: plt.ylabel('f(x)')
    i = i + 1
    
plt.tight_layout()

```

## Método da Bisecção

```python
def biseccao(func,xi,xs,tol):
    
    Er_r = [np.inf]
    y = [np.NaN]
    xr = [xi]
    i = 0
    while abs(Er_r[i]) > tol:
        if func(xi)*func(xs) < 0:
            xr.append((xi + xs)/2)
            y.append(func(xr[i+1]))
            Er_r.append(((xr[i+1] - xr[i])/xr[i+1])*100)

            if func(xi)*func(xr[i+1]) < 0:
                xs = xr[i+1]
            else:
                xi = xr[i+1]
        else:
            print(f'Não existe raiz entre o intervalo xi = {xi} e xs = {xs}')
            break

        i = i + 1

    return i, Er_r, xr
```

## Método da Falsa Posição

```python
def falsa_posição(func,xi,xs,tol):
    
    Er_r = [np.inf]
    y = [np.NaN]
    xr = [xi]
    i = 0
    while abs(Er_r[i]) > tol:
        if func(xi)*func(xs) < 0:
            xr.append(xs - func(xs)*(xi - xs)/(func(xi)-func(xs)))
            y.append(func(xr[i+1]))
            Er_r.append(((xr[i+1] - xr[i])/xr[i+1])*100)

            if func(xi)*func(xr[i+1]) < 0:
                xs = xr[i+1]
            else:
                xi = xr[i+1]
        else:
            print(f'Não existe raiz entre o intervalo xi = {xi} e xs = {xs}')
            break

        i = i + 1

    return i, Er_r, xr
```

### Equação 1

Implementando os métodos de Bisecção e Falsa Posição 

```python
max_iter_bis, Erro_rel_bis, raiz_bis = biseccao(eq_1,5,10,0.0001)
max_iter_falp, Erro_rel_falp, raiz_falp = falsa_posição(eq_1,5,10,0.0001)

bis_dict = {'Iterações': range(max_iter_bis+1),'Erro_Relativo': Erro_rel_bis}
falp_dict = {'Iterações': range(max_iter_falp+1),'Erro_Relativo': Erro_rel_falp}

bis_df = pd.DataFrame(bis_dict)
falp_df = pd.DataFrame(falp_dict)
```

```python
bis_df.to_csv('Q1_Resultados_biseccao.csv', index=False)
falp_df.to_csv('Q1_Resultados_falpos.csv', index=False)
```

```python
plt.figure()
plt.plot(range(max_iter_bis+1),Erro_rel_bis,label='Bisecção')
plt.plot(range(max_iter_falp+1),Erro_rel_falp,label='Falsa Pos.')
plt.legend()
plt.title('Erro relativo - Questão 1')
plt.xlabel('Iterações [-]')
plt.ylabel('Erro relatiivo [%]')
plt.show()
```
